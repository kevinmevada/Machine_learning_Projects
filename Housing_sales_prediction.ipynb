Importing important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

Read the excel file of housing data

```python
housing_data = pd.read_excel(r'/HOUSEPRED.xlsx')
housing_data.head(30)
```

Removed Null values and replaced it with respective Median

```python
import math
median_bedroom= math.floor(housing_data.bedrooms.median())
median_bathrooms= math.floor(housing_data.bathrooms.median())
median_sqft_living= math.floor(housing_data.sqft_living.median())

housing_data['bedrooms'] = housing_data['bedrooms'].fillna(median_bedroom)
housing_data['bathrooms'] = housing_data['bathrooms'].fillna(median_bathrooms)
housing_data['sqft_living'] = housing_data['sqft_living'].fillna(median_sqft_living)

housing_data.head(30)
```

Made the regression_model for prediction using LinearRegression

```python
regression_model = linear_model.LinearRegression()
regression_model.fit(housing_data[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','condition','sqft_above','sqft_basement','yr_built']],housing_data.price)
```

Predicted the data from the Excel

```python
predict_dataset= pd.read_excel(r'/housepredd.xlsx')
prediction = regression_model.predict(predict_dataset)
prediction
```

Added predicted data and converted it into excel

```python
predict_dataset['Prices'] = prediction
predict_dataset.to_excel('final_prediction.xlsx', index=False)
```
